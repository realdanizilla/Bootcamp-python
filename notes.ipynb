# aula 11 Introducao a Orientacao a objetos

orientacao a objeto nao é tao necessaria em projetos mais simples e locais

OOP criar abstracoes e entender como objetos

usado para aplicar um mesmo codigo para que ele seja reproduzido em varias situacoes e interacoes com outros objetos

quando temos multiplos desenvolvedores, padroes de escrita, código sendo feito todo dia, começa a afazer sentido pensar em classes, pois aí fica mais reciclável, e future-proof

funcao dentro de classe é chamada de método

para pensar em classes vamos pensar em entidades e pecas fundamentais que temos na necessidade do negocio ou sistema e pensar em o que elas tem em comum para criar uma abstacao

exemplo temos cao, vaca e passaro, o que esses tem em comum? sao animais, entao animais poderia ser uma classe com alguns atributos e métodos que seriam comuns a todos eles

ex leitura e salvamento de arquivos, se tenho arquivos em varios formatos, poderia ter uma classe 'recebedor de arquivos' que tem métodos como:
-  receber_arquivo
- transformar_arquivo_df
- salvar_arquivo
- filtrar_arquivo

ai quando vier um novo desenvolvedor ou novo formato de arquivo, desenvolvemos já seguindo uma lógica facil de integrar no resto do código

classe abstrata é quando crio uma classe e seus métodos mas náo entro no detalhe de cada método, porém cada nova instancia dessa classe precisara definir esses métodos específicos para cada instancia da classe. Define o fluxo que a nova forma de dados vai ter que passar

posso definir métodos que realmente seriam comuns a todas as classes no futuro e entao uma nova instancia da classe se beneficia desse método ao inves de ter que desenvolver ele do zero e diferente da logica das demais instancias

funcao vc chama, ele tem um input, retorno edepois ela morre. A classe nao, ela persiste na instancia

herdar o método pode ser diferente de herdar a implementacao do método, caso necessário ou conveninente

a classe vai te ajudar amanha e nao hoje, facilita o trabalho do desenvolvedor, que é caro

tem a ver com a experiencia do desenvolvimento

## Aula 20 CRUD na prática

docker compose faz a orquestracao - zipar os arquivos de backend, frontend e db

qual a diferença entre models e schema?
- Tabela do banco é o model
- schema é a validacao, que depende da operacao (funcao), ou seja pode ser diferente da tabela do banco( exemplo em uma funcao de create nao vamos precisar de id e nem de created date, porem quando eu for ler ele vai ser igua la tabela do banco, em um update será ainda mais diferente)
 - portanto temos um schema para cada operacao para validar a informacao que está chegando
- validacoes sao mais específicas que os models
- model é a representacao do banco
- schema é a representacao das interacoes com o usuario
